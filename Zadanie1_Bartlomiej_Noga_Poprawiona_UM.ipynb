```python
import random
```

```python
def policy(observation, t):

    actions = {'south' : 0, 'north' : 1, 'east' : 2, 'west' : 3, 'pick-up' :4, 'drop-off' : 5}
   
    point = observation / 4

    if 100 < point < 200:
       return actions['west']
    elif point < 80:
       return actions['south']
    elif point > 300: 
       return actions['north']
    elif 200 < point < 300:
        return actions['west']
    else:
       return actions['east']

'''
Dosyć cięzko mi wymyślić jakiś rozsądny algorytm samodzielnie ponieważ obsługa instrukcji warunkowych powoduje blokowanie się taxi 
'''

   
```

```python
observation = env.reset()

for t in range(199):
    
    #Akcje wybierane według polityki policy:
    action = policy(observation, t)
    
    #Akcja zostaje wykonana, stan środowiska aktualizuje się:
    observation, reward, done, _ = env.step(action)
    print("Nr of iteration {}| position {}| reward {}| is done {}|".format(t,observation, reward, done))
    env.render()

```
