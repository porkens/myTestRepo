```python
import numpy as np
import gym
import random
```

```python
env = gym.make("Taxi-v3").env
env.render()
```

```python
action_size = env.action_space.n
print("Action size:", action_size) # Przestrzeń akcji

state_size = env.observation_space.n 
print("State size:", state_size) # Przestrzeń stanów
```

```python
qtable = np.zeros((state_size, action_size)) # Tworzymy tabelę Q, którą wypełniamy zerami, gdzie liczba wierszy jest równa liczbie stanów(w tym przypadku 500), a liczba kolumn jest równa liczbie akcji(6 kolumn)
print(qtable) # Tabela Q
```

```python
total_episodes = 10000
total_test_episodes = 10
max_steps = 99

learning_rate = 0.7
gamma = 0.618

epsilon = 1.0
max_epsilon = 1.0
min_epsilon = 0.01
decay_rate = 0.01
```

```python
for episode in range(total_episodes):
  state = env.reset()
  step = 0
  done = False

  for step in range(max_steps):
    exp_exp_tradeoff = random.uniform(0,1)

    if exp_exp_tradeoff > epsilon:
      action = np.argmax(qtable[state,:])

    else:
      action = env.action_space.sample()

    new_state, reward, done, info = env.step(action)

    qtable[state, action] = qtable[state, action] + learning_rate * (reward + gamma * np.max(qtable[new_state, :]) - qtable[state, action])

    state = new_state

    if done == True:
      break
  
  episode += 1

  epsilon = min_epsilon + (max_epsilon - min_epsilon) * np.exp(-decay_rate*episode)
```

```python
env.reset()
rewards = []

for episode in range(total_test_episodes):
  state = env.reset()
  step = 0
  done = False
  total_rewards = 0
  print("##################")
  print("   Zdarzenie:", episode)
  print("##################")

  for step in range(max_steps):
    env.render()
    action = np.argmax(qtable[state, :])

    new_state, reward, done, info = env.step(action)

    total_rewards += reward

    if done:
      rewards.append(total_rewards)
      print("Wynik:", total_rewards)
      break
    state = new_state
env.close()
print("\nŚrednia: " + str(sum(rewards)/total_test_episodes))
```
